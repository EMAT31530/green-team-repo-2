```python
# This code looks at using gradient boosting. 

# Boosing is learning from previous mistakes made. i.e. too far off with regression. 

# https://towardsdatascience.com/basic-ensemble-learning-random-forest-adaboost-gradient-boosting-step-by-step-explained-95d49d1e2725
```

Gradient boosting is better than Adaboost for our application as it is less senstitive to outliers.

Adaboost minminses the exponential loss function that can make the algorithm sensitive to outliers
In Adaboost, errors are identified by high-weight data points

Gradient boost - any differentiable loss function can be utilised. This means that it is more robust. 
In Gradient Boosting, errors of existing weak learners are identified by gradients. 

Gradient boosting is better than random forest for this application. The key differences between the two algorithms are how the trees are built and how the results are combined. 

In gradient boosting, the trees are built one at a time. This means that the errors made in the tree before inform the next, or this can be thought of as introducing a weak learner to improve the shortcummings of the existing weak learner. 
When the results are combined in random forest, an average is taken. Gradient boosting combines the results along the way. 

Random forests peform better where there is a lot of noise in the data, as it reduces overfitting. Gradient boosting works well in this application as the data is unbalanced with a number of outliers. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.feature_selection import SelectFromModel
```

```python
nutrition = pd.read_pickle("./Nutrition_Full_Features.pkl")
#nutrition.drop(nutrition.columns[[]], axis=1, inplace=True) 

# removing the brand, and food type columns for data cleaning. 
nutrition
```

```python
nutrition = nutrition.dropna() # dropping NaN values. This ensures random forests can accept data. 
nutrition
```

```python
nutrition.dtypes # checking types of data
```

```python
X = nutrition.iloc[:, 11:].astype('float64').dropna() 
# pulling out the features data
X
```

```python
y = nutrition.iloc[:, 4:5].astype('float64').dropna()
# pulling out the price data
y
```

```python
# flattening the array for use in the random forest algorithm
y = np.ravel(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train_validation, X_test, y_train_validation, y_test = train_test_split(X,y,test_size=0.5)
# splitting the data into train and test data. 
#size =0.2 means 80% data is training data, 20% testing
```

```python
# splitting the training data into validation and training data. 
X_train, X_validation, y_train, y_validation = train_test_split(X_train_validation,y_train_validation,test_size=0.15)
```

```python
# Implementing GradientBoostingRegressor here
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(n_estimators=75)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
#automatically select features 
sel = SelectFromModel(model)
sel.fit(X_train, y_train)
```

```python
#seeing which features are important 
#returns an array of trues when features are more important than mean 
# false when features are less important than mean importance 
sel.get_support()

# make a list and count selected features 

selected_feat = X_train.columns[(sel.get_support())]
#return the number of features selected by the random forest 
len(selected_feat)
```

```python
#get the name of feautures selected 
print(selected_feat)
```

```python
Importances_idx = model.feature_importances_.argsort()

Features = X_train.columns[Importances_idx]
Importances = model.feature_importances_[Importances_idx]
```

```python
# checking the outputs 
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
Mean_Absolute_Error = []
Mean_Squared_Error = []
Root_Mean_Squared_Error = []
```

```python
# creating a for loop to evaluate the hyperperameters of the model. Initally with n_estimators
# implementing the gradient boosting regression on the for loop
for i in range (1, 200, 5):
    from sklearn.ensemble import GradientBoostingRegressor
    model = GradientBoostingRegressor(n_estimators=i)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mae = metrics.mean_absolute_error(y_test, y_pred)
    mse = metrics.mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
    
    Mean_Absolute_Error.append(mae)
    Mean_Squared_Error.append(mse)
    Root_Mean_Squared_Error.append(rmse)
    
    
```

```python
# plotting the different metrics to see the effect of changing the number of trees
import matplotlib.pyplot as plt
n_trees = list(range(1, 200, 5))

f = plt.figure(1)

plt.subplot(311)
plt.plot(n_trees, Mean_Absolute_Error)
plt.ylabel('Mean Absolute Error')
plt.xlabel('Number of Trees')
plt.show()
plt.subplot(312)
plt.plot(n_trees, Mean_Squared_Error)
plt.ylabel('Mean_Squared_Error')
plt.xlabel('Number of Trees')
plt.show()
plt.subplot(313)
plt.plot(n_trees, Root_Mean_Squared_Error)
plt.ylabel('Root_Mean_Squared_Error')
plt.xlabel('Number of Trees')
plt.show()
```
