```python
import numpy as np # numbers
import pandas as pd #data
import matplotlib.pyplot as plt #graphs
```

```python
#data in order of newest additions at the top
price_data = pd.read_csv(r'/Users/rachaelgreen/Desktop/Third_year_maths/Intro_to_ai/GitHub/ai-group-project-green-team/Prices_of_food/kosherfooddataset.csv', encoding='latin1')
price_data
# encoding='latin1' i used this to fix an error is that ok?
```

```python
#dropping the columns that we definitely dont need
price_data = price_data.drop(columns=['Enabled', 'Stock', 'CreatedÊ' ,'Unnamed: 8', 'Unnamed: 9']) 
price_data
```

```python
#dropping the collumns we probably dont need
price_data = price_data.drop(columns=['ID', 'Supplier', 'Manufacturer'])
price_data
```

```python
#get rid of any duplicates
#removes duplicate rows based on all columns
price_data = price_data.drop_duplicates()
price_data
```

```python
#To remove duplicates on specific column(s), use subset, and keep last occurences, use keep.
price_data = price_data.drop_duplicates(subset=['Name'], keep='first') #newest price
price_data
```

```python
#drop any row where at least one element is missing
price_data = price_data.dropna()
price_data
```

```python
#re-number index
price_data = price_data.reset_index()
price_data
```

```python
price_data = price_data.drop(columns=['index'])
price_data
```
